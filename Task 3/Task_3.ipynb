##  Yearwise Income Bar Plot and Line Plot

```python
column_names
```

```python
yrs = list(df[column_names[0]].values)
print(yrs)
```

```python
years = sorted(list(set(df[column_names[0]].values)))
years
```

```python
income_2008 = sum(df["income"].values[0:yrs.count(years[0])])
income_2011 = sum(df["income"].values[yrs.count(years[0]):yrs.count(years[0])+yrs.count(years[1])])
income_2014 = sum(df["income"].values[yrs.count(years[0])+yrs.count(years[1]):yrs.count(years[0])+yrs.count(years[1])+yrs.count(years[2])])
income_2017 = sum(df["income"].values[yrs.count(years[0])+yrs.count(years[1])+yrs.count(years[2]):yrs.count(years[0])+yrs.count(years[1])+yrs.count(years[2])+yrs.count(years[3])])
income_2020 = sum(df["income"].values[yrs.count(years[0])+yrs.count(years[1])+yrs.count(years[2])+yrs.count(years[3]):yrs.count(years[0])+yrs.count(years[1])+yrs.count(years[2])+yrs.count(years[3])+yrs.count(years[4])])

yearwise_income = [income_2008, income_2011, income_2014, income_2017, income_2020]
yearwise_income
```

```python
# Bar Graph
plt.bar(years, yearwise_income, color="orange")
plt.xlabel("Years", fontdict=font2)
plt.ylabel("Income", fontdict=font2)
plt.title("Yearwise Income", fontdict=font1)
plt.grid()
plt.show()
```

```python
# Line Plot
plt.plot(years, yearwise_income, linestyle="dashed",c="green")
plt.xlabel("Years", fontdict=font2)
plt.ylabel("Income", fontdict=font2)
plt.title("Yearwise Income", fontdict=font1)
plt.grid()
plt.show()
```
