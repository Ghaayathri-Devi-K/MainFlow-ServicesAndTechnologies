## Data Cleaning and Data Manipulation

```python
new_df = df.copy()
```

```python
new_summary = new_df.describe()
```

```python
column_names = new_df.columns
print(column_names[0:len(column_names)-1])
```

### Handling Missing Values

```python
for i in column_names[1:]:
   X = new_df[i].mean()
   new_df[i].fillna(X, inplace=True)
```

```python
new_df.info()
```

### Data Filtering

```python
for j in column_names[1:]:
  threshhold = new_summary[j]["75%"]
  filtered_rows = new_df[j] <= threshhold
  filtered_df = new_df.loc[filtered_rows]

filtered_df.info()
```

## Calculating Summary Statistics

```python
filtered_df.describe()
```

```python
# Frequency Table

for t in column_names[1:]:
  freq_tbl = filtered_df[t].value_counts()
  print(freq_tbl)
  print("---------------")
```
