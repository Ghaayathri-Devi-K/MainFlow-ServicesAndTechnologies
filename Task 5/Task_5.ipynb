```python
# Checking for null values
df.notnull().sum()
```

Hence, there are no null values in the dataset

```python
# Checking for duplicates
df.nunique() # Gives the unique categories in each feature
```

```python
# Getting the feature names
column_names = list(df.columns)
print(column_names)
```

```python
categorical_columns = df.columns[df.dtypes == 'object']
print(categorical_columns)
```

```python
df = pd.get_dummies(df, columns=categorical_columns, drop_first=True)
```

```python
# Creating training and testing split

from sklearn.model_selection import train_test_split
```

```python
columns_to_keep = df.columns[df.columns != 'target'].tolist()
X_train, X_test, y_train, y_test = train_test_split(df[columns_to_keep], df['target'], test_size=0.25, random_state=1)
```

```python
# Standardize the dataset

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train_std = sc.fit_transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
# Performing Principal Component Analysis
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train_std)
X_test_pca = pca.transform(X_test_std)
```
